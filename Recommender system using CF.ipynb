```python
import pandas as pd
from surprise import Dataset
from surprise import Reader
```

```python
#Movielense 100k dataset
```

```python
movielense = Dataset.load_builtin('ml-100k')

train_data = movielense.build_full_trainset()
```

```python
#view data through Pandas
```

```python
raw_data = movielense.raw_ratings

df = pd.DataFrame(raw_data,columns=['user_id','item_id','rating','timestemp'])

print(df.head())
```

```python
#KNNWithMeans
```

```python
from surprise import KNNWithMeans

sim_options = {"name":"cosine", "user_based":False,}

recommender = KNNWithMeans(sim_options = sim_options)
```

```python
recommender.fit(train_data)
```

```python
pred = recommender.predict(200,100)
print(pred)
```

```python
#hyperparameter Tuning
```

```python
from surprise.model_selection import GridSearchCV

sim_options = {"name":["msd","cosine"],"min_support":[3,4,5],"user_based":[False,True],}

param_grid = {"sim_options":sim_options}

gs = GridSearchCV(KNNWithMeans,param_grid,measures=["rmse","mae"],cv=3)

gs.fit(movielense)
```

```python
print(gs.best_score["rmse"])
print(gs.best_params["rmse"])
```

```python
#SVD
```

```python
from surprise import SVD
```

```python
param_grid = {"n_epochs":[5,10],"lr_all":[0.002,0.005],"reg_all":[0.4,0.6]}

gs_svd = GridSearchCV(SVD, param_grid,measures = ["rmse","mae"],cv=3)

gs_svd.fit(movielense)


```

```python
print(gs_svd.best_score["rmse"])
print(gs_svd.best_params["rmse"])
```
